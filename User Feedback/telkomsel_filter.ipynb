```python
# Define the keywords to filter the reviews
keywords = ['telkomsel', 'mytelkomsel', 'tsel', 'mytsel', 'my telkomsel', 'telkmsel', 'telkomsel']

# Filter the rows that contain any of the keywords (case insensitive)
# df_new = df[df['content'].str.contains('|'.join(keywords), case=False, na=False)] #if using Google Play
df_new = df[df['review'].str.contains('|'.join(keywords), case=False, na=False)] #if using Appstore

# Display the filtered DataFrame to the user
# import ace_tools_open as tools; tools.display_dataframe_to_user(name="Filtered GooglePlay Reviews", dataframe=df_new) #if using Google Play
import ace_tools_open as tools; tools.display_dataframe_to_user(name="Filtered Appstore Reviews", dataframe=df_new) #if using Appstore
```
