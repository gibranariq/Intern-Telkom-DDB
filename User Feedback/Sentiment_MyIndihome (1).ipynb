```python
model = genai.GenerativeModel('gemini-1.5-flash-latest') #15RPM #1500RPD
```

```python
import time

def sentiment_predict(df):
    total_rows = len(df)

    for i in range(total_rows):
        try:
            prediction = model.generate_content(f"Produk kita bernama MyIndihome, Anda adalah ahli bahasa, yang pandai mengklasifikasikan sentimen review pelanggan terhadap produk kita ke dalam label Positif, Negatif,dan Netral. Bantu saya mengklasifikasikan review pelanggan menjadi Positif, Negatif, atau Netral pada kalimat berikut, '{df.loc[i, 'review']}'. Hanya berikan jawaban dengan 1 kata antara Positif, Negatif, atau Netral.")
            df.loc[i, 'sentiment'] = prediction.text
        except Exception as e:
            print(f"[{i}]> Error: {e}")
            df.loc[i, 'sentiment'] = "Error"  # Tampilkan pesan error jika respons tidak valid

        if (i + 1) % 15 == 0:
            time.sleep(60)  # Istirahat 1 menit (60 detik)

    return df
```

# Topic Prediction

```python
import time

def topic_predict(df):
    total_rows = len(df)

    for i in range(total_rows):
        try:
            prediction = model.generate_content(f"Produk kita bernama MyIndihome, Anda adalah ahli dalam pengkategorian, yang pandai mengklasifikasikan review pelanggan terhadap produk kita ke dalam salah satu kategori Usability & Reliability Apps (segala permasalahan/kendala berkaitan dengan penggunaan dan kehandalan aplikasi seperti aplikasi error,bug,lamat dan sebagainya), Customer Service (segala permasalahan/kendala berkaitan dengan layanan customer yang diberikan seperti lambatnya penanganan laporan/gangguan, pelayanan petugas tidak memuaskan,dan sebagainya), Kualitas Internet (segala permasalahan/kendala berkaitan dengan kualitas internet seperti gangguan internet dalam kurun waktu, internet lambat, tidak stabil,dan sebagainya), Payment (segala permasalahan/kendala berkaitan dengan pembayaran seperti tagihan tetap muncul walaupun sudah dibayarkan, dan sebagainya), Unidentified (Review tidak jelas, tidak memiliki konteks yang cukup untuk dikategorikan). Bantu saya mengklasifikasikan review pelanggan kedalam kateogri tersebut berdasarkan review berikut, '{df.loc[i, 'review']}'. Hanya berikan jawaban dengan tepat 1 kategori antara [Usability & Reliability Apps, Customer Service,Kualitas Internet, Payment, Unidentified].")
            df.loc[i, 'Kategori'] = prediction.text
        except Exception as e:
            print(f"[{i}]> Error: {e}")
            df.loc[i, 'Kategori'] = "Error"  # Tampilkan pesan error jika respons tidak valid

        if (i + 1) % 15 == 0:
            time.sleep(60)  # Istirahat 1 menit (60 detik)

    return df
```
