```python
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown


def to_markdown(text):
    text = text.replace('•', '  *')
    return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
model = genai.GenerativeModel('gemini-1.5-flash-latest') #15RPM #1500RPD
```

```python
import time

pilihan_topik = ['Pengalaman Pelanggan (Customer Experience)',
                 'Fitur Produk & Layanan (Product & Service Features)',
                 'Proses Pengiriman & Jaminan (Delivery & Assurance Process)',
                 'Strategi Pemasaran / Go To Market (GTM)',
                 'Perbandingan dengan Kompetitor (Competitor Analysis)',
                 'Harga & Pola Pembayaran (Price & Payment Pattern)',
                 'Lainnya (Others)']

def topic_predict(df):
    total_rows = len(df)

    for i in range(total_rows):
        try:
            prediction = model.generate_content(f"Anda adalah seorang data scientist dan ahli dalam membuat topik data feedback pelanggan dari produk-produk digital. Bantu saya mendefinisikan feedback pelanggan ke dalam topik yang tepat dan sesuai berdasarkan feedback berikut, '{df.loc[i, 'List Feedback']}'. Hanya berikan 1 jawaban yang tepat dari pilihan topik berikut {pilihan_topik}. Berikan langsung jawabannya saja.")
            df.loc[i, 'Topik'] = prediction.text.replace(' \n','')
        except Exception as e:
            print(f"[{i}]> Error: {e}")
            df.loc[i, 'Topik'] = "Error"  # Tampilkan pesan error jika respons tidak valid

        if (i + 1) % 15 == 0:
            time.sleep(60)  # Istirahat 1 menit (60 detik)

    return df
```
