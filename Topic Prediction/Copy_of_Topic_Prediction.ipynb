```python
model = genai.GenerativeModel('gemini-1.5-flash-latest') #15RPM #1500RPD
```

```python
import time

def topic_predict(df):
    total_rows = len(df)

    for i in range(total_rows):
        try:
            prediction = model.generate_content(f"Produk kita bernama IndiHome, Anda adalah ahli bahasa, yang pandai mengklasifikasikan topik ulasan pelanggan ke dalam 6 kategori yaitu: Layanan, Kecepatan, Migrasi, Harga, Login, dan Unidentified. Bantu saya mengklasifikasikan ulasan pelanggan menjadi Layanan, Kecepatan, Harga, Migrasi, Login, dan Unidentified pada kalimat berikut, '{df.loc[i, 'clean_text']}'. Hanya berikan jawaban dengan 1 kata antara Layanan, Kecepatan, Harga, Migrasi, Login, Unidentified.")
            df.loc[i, 'topik'] = prediction.text
        except Exception as e:
            print(f"[{i}]> Error: {e}")
            df.loc[i, 'topik'] = "Error"  # Tampilkan pesan error jika respons tidak valid

        if (i + 1) % 15 == 0:
            time.sleep(60)  # Istirahat 1 menit (60 detik)

    return df
```
